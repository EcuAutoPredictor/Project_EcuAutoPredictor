```python
import pandas as pd

# Carga el DataFrame. Cambia 'ruta_al_archivo.csv' por la ruta a tu archivo de datos.
df = pd.read_csv('../scrapers/data_1001carros.csv')

# Extrae la primera letra de la columna "Placa" y la convierte a mayúscula
df['Letra'] = df['Placa'].str[0].str.upper()

# Define el diccionario de mapeo para la variable "Provincia"
# Aquí puedes agregar las otras correspondencias. Por ejemplo, para 24 provincias:
provincia_dict = {
    'A': 'AZUAY',
    'B': 'BOLIVAR',
    'U': 'CAÑAR',
    'C': 'CARCHI',
    'H': 'CHIMBORAZO',
    'X': 'COTOPAXI',
    'O': 'EL ORO',
    'E': 'ESMERALDAS',
    'W': 'GALAPAGOS',
    'G': 'GUAYAS',
    'I': 'IMBABURA',
    'L': 'LOJA',
    'R': 'LOS RIOS',
    'V': 'MORONA SANTIAGO',
    'N': 'NAPO',
    'Q': 'ORELLANA',
    'S': 'PASTAZA',
    'P': 'PICHINCHA',
    'Y': 'SANTA ELENA',
    'J': 'SANTO DOMINGO',
    'K': 'SUCUMBIOS',
    'T': 'TUNGURAHUA',
    'Z': 'ZAMORA CHINCHIPE'
}

# 3. Crea la nueva variable "Provincia" aplicando el mapeo de la letra extraída
df['Provincia'] = df['Letra'].map(provincia_dict)

# 4. Elimina las columnas "Placa" y la auxiliar "Letra"
df.drop(['Placa', 'Letra'], axis=1, inplace=True)


# 6. Reordena las columnas del DataFrame según el orden deseado
orden_columnas = [
    "Marca", 
    "Modelo", 
    "Precio", 
    "Provincia", 
    "Año", 
    "Kilometraje", 
    "Transmisión", 
    "Dirección", 
    "Motor", 
    "Tracción", 
    "Color", 
    "Combustible", 
    "Enlace"
]

# Asegúrate de que todas estas columnas existan en tu DataFrame; si alguna no está, se generará un error.
df = df[orden_columnas]

# 7. Guarda el DataFrame modificado en un nuevo archivo CSV
df.to_csv('data_1001carros_C.csv', index=False, encoding='utf-8')

print("El archivo modificado se ha guardado exitosamente en 'data_1001carros_C.csv'")
```

```python
import pandas as pd

# Carga el DataFrame (modifica la ruta según corresponda)
df = pd.read_csv('archivo_unido.csv')

# Filtra los registros eliminando aquellos que tengan "No encontrado" en "Precio" y en "Provincia"
df_filtrado = df[(df['Precio'] != "No encontrado") & (df['Provincia'] != "NO ENCONTRADO")]




df_filtrado.loc[:, 'Motor'] = pd.to_numeric(df_filtrado['Motor'], errors='coerce') * 1000



# Reemplaza en la columna "Tracción" los valores "4 x 2" por "4x2" y "4 x 4" por "4x4"
df_filtrado['Tracción'] = df_filtrado['Tracción'].replace({"4 x 2": "4x2", "4 x 4": "4x4"})


# Muestra la cantidad de registros antes y después del filtrado
print("Registros originales:", df.shape[0])
print("Registros después del filtrado:", df_filtrado.shape[0])

# Guarda el DataFrame filtrado en un nuevo archivo CSV
df_filtrado.to_csv('archivo_filtrado.csv', index=False, encoding='utf-8')

print("El archivo filtrado se ha guardado exitosamente en 'archivo_filtrado.csv'")
```

```python
import pandas as pd

# Cargar el DataFrame (Asegúrate de que el archivo existe en tu directorio de trabajo)
df = pd.read_csv('archivo_unido_FINAL2.csv')

# Agrupar los modelos por marca
marcas_modelos = df.groupby("Marca")["Modelo"].unique().to_dict()

# Mostrar los datos en la terminal
for marca, modelos in marcas_modelos.items():
    print(f"Marca: {marca}")
    print(f"Modelos: {', '.join(modelos)}")
    print("-" * 50)

# Guardar el resultado en un archivo CSV
df_marcas_modelos = pd.DataFrame([(marca, ", ".join(modelos)) for marca, modelos in marcas_modelos.items()],
                                  columns=["Marca", "Modelos"])
df_marcas_modelos.to_csv("marcas_y_modelos.csv", index=False)

print("\n✅ Archivo 'marcas_y_modelos.csv' generado con éxito.")
```
