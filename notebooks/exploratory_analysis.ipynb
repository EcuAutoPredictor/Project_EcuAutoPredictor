```python

# 游늷 Importar librer칤as necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 游늷 Definir ruta del dataset limpio
data_path = "../data/final/sri_autos_final.csv"
```

```python
# 游늷 Cargar el dataset
df = pd.read_csv(data_path, delimiter=';', encoding="latin-1", low_memory=False)
```

```python
# 游늷 Mostrar informaci칩n general del dataset
print("游늷 Informaci칩n del dataset:")
print(df.info())
```

```python

# 游늷 Descripci칩n estad칤stica de las variables num칠ricas
print("\n游늷 Estad칤sticas descriptivas:")
print(df.describe())
```

```python
# 游늷 Histograma de "avaluo"
plt.figure(figsize=(10, 5))
sns.histplot(df["avaluo"], bins=30, kde=True, color="blue")
plt.xlabel("Avaluo ($)")
plt.ylabel("Cantidad de Autos")
plt.title("Distribuci칩n de Aval칰os de Veh칤culos en Ecuador")
plt.show()
```

```python
# 游늷 Histograma de "cilindraje"
plt.figure(figsize=(10, 5))
sns.histplot(df["cilindraje"], bins=30, kde=True, color="red")
plt.xlabel("Cilindraje")
plt.ylabel("Cantidad de Autos")
plt.title("Distribuci칩n del Cilindraje de Veh칤culos")
plt.show()
```

```python

# 游늷 Relaci칩n entre "avaluo" y "year_modelo"
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["year_modelo"], y=df["avaluo"], alpha=0.5)
plt.xlabel("A침o del Modelo")
plt.ylabel("Avaluo ($)")
plt.title("Relaci칩n entre A침o del Modelo y Aval칰o")
plt.show()
```

```python
# 游늷 Relaci칩n entre "cilindraje" y "avaluo"
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["cilindraje"], y=df["avaluo"], alpha=0.5)
plt.xlabel("Cilindraje")
plt.ylabel("Avaluo ($)")
plt.title("Relaci칩n entre Cilindraje y Aval칰o")
plt.show()
```
